# Quick example - inference with trained models

This notebook demonstrates how you can read the directories saved by the model training script `gim-cv/bin/train_segmentalist.py`, 
compare model performance and load the weights of a trained model, then use this for inference.

```python
import os
import sys
import yaml

from pathlib import Path

import numpy as np
import dask.array as da
import matplotlib.pyplot as plt

import gim_cv.config as cfg
from gim_cv.models.segmentalist import Segmentalist

# the model saving/loading utility functions live in the utils.py module in bin with the scripts
sys.path.append('../../bin')
from bin import utils
```

First we load the metadata saved in the "run_params.yml" of every saved model using `utils.collate_run_data`:

```python
print(os.listdir('../../saved_models/ebs_trained_models'))
```

```python
df_segm_trained = utils.collate_run_data(
    Path('../../saved_models/ebs_trained_models/Segmentalist'), # this should match the "models_dir" arg used in train_segmentalist.py
    model_name='Segmentalist' # this will search for run data saved in directories beginning "Segmentalist"
)
```

This constructs a `pd.DataFrame` with all of the parameters passed to the script (including model hyperparameters, training data etc.)

```python
df_segm_trained.head()
```

We can throw away some args that we don't care about for comparing them to make it a bit more compact and sort by validation loss:

```python
df_segm_trained = df_segm_trained.drop(
    ['seed', 'save_to_cache'] + 
    [c for c in df_segm_trained.columns if any(
        i in c for i in (
            '_train',  'ocp', 'fancy_augs', 'dump_test_data',
            'dump_first_batches', 'channel_expansion_factor', 'balanced_oversample'
        )
    )],
    axis=1
).sort_values(by='lowest_val_loss')
```

Select which run parameters to view (the hyperparameters etc. of interest that we have been experimenting with):

```python
# we can select a sub
view_cols = ['lowest_loss', 'lowest_val_loss', 'cardinality',
             'csag', 'sag', 'overlapping_tiles', 'deep_supervision',
             'pyramid_pooling', 'encoder_cbam', 'decoder_cbam',
             'val_frac', 'lr_init', 'lr_min']
```

We can query this to look for models with a certain loss and training data, and then see an overview of how the relevant hyperparameters performed.

Here we look for models trained with the weighted binary cross entropy (`wbce_adaptive`) loss function using all the 40cm dataset `col_med_19_40cm`:

```python
df_segm_trained.query('datasets == "col_med_19_40cm" and loss_fn == "wbce_adaptive"')[view_cols + ['uuid4']]
```

Similarly for the Tversky loss (`tversky_loss`) function:

```python
df_segm_trained.query('datasets == "col_med_19_40cm" and loss_fn == "tversky_loss"')[view_cols + ['uuid4']]
```

We can select a certain row for which we want to load the weights:

```python
row = df_segm_trained.query('datasets == "col_med_19_40cm" and loss_fn == "wbce_adaptive"').iloc[0]
row
```

Now we can load the validation data used in that training run so we can look at the results of the model here (the validation arrays are saved by the `train_segmentalist.py` script):

```python
X_val = np.load(row.X0_val) # takes a wee minute
```

We can load the model weights just by passing the row of interest to `Segmentalist.load_from_metadata`:

```python
model = Segmentalist.load_from_metadata(row=row)
# shouldn't have to run these lines, LRM will look at the method above and fix so weight loading is included properly...
# workaround which fixes weight loading; "prime" model on something of the right shape
model(np.random.rand(1, 256, 256, 3))
model.load_weights(row.lowest_val_loss_ckpt) # takes a wee minute
```

Now predict the first batch of 4 (note: have to normalise RGB by 1/255.):

```python
y_val_pred = model(X_val[:4]/255.)
```

```python
ix = 2
fig, axs = plt.subplots(1, 2, figsize=(16, 8))
axs[0].imshow(X_val[ix])
axs[1].imshow(y_val_pred[ix])
```

# Run inference

```python
from gim_cv.datasets import list_datasets
```

```python
print(list_datasets(skip_missing_files=True))
```

```python
inference_window_size = 896

# load again with new patch shape
model = Segmentalist.load_from_metadata(row=row)
# shouldn't have to run these lines, LRM will look at the method above and fix so weight loading is included properly...
# workaround which fixes weight loading; "prime" model on something of the right shape
model(np.random.rand(1, inference_window_size, inference_window_size, 3))
model.load_weights(row.lowest_val_loss_ckpt) # takes a wee minute
```

Pick a dataset on which to run inference:

```python
from gim_cv.datasets import get_dataset

ds = get_dataset('col_med_19_test_40cm')
ids = ds.load_inference_data(inference_window_size=inference_window_size)
ids.prepare()
```

Run inference:

```python
# prepare inference with model
# choose where to write files
out_dir = cfg.volumes_data_path / Path(f'datasets/medellin_oxford_inria/segmentation_outputs_{row.uuid4}')

ids.schedule_inference(
    model,
    output_directory=out_dir
)

ids.write_mask_rasters(overwrite=True)
```
